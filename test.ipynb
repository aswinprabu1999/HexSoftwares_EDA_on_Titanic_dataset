```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.get_dataset_names()
df=sns.load_dataset("titanic")
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.shape
df.nunique()
df.describe(include='O')
```

```python
# Check missing values percentage
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)

# Drop columns with too many missing values (e.g., 'deck')
df.drop(columns=["deck"], inplace=True)

# Fill missing 'age' with median
df['age'].fillna(df['age'].median(), inplace=True)

# Fill missing 'embarked' with the most common value
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns="embark_town",inplace=True)
```

```python
# Countplot for survival
sns.countplot(data=df, x="survived", palette="Set2")
plt.title("Survival Distribution")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
# Barplot for survival rate by gender
sns.barplot(data=df, x="sex", y="survived", palette="muted",legend=True)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
# Age distribution plot
sns.histplot(df['age'], kde=True, bins=30, color="red")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
# Survival rate by passenger class
sns.barplot(data=df, x="pclass", y="survived", palette="coolwarm")
plt.title("Survival Rate by Class")
plt.show()
```

```python
# Fare distribution plot
sns.boxplot(data=df, x="pclass", y="fare", palette="viridis")
plt.title("Fare Distribution by Class")
plt.show()
```

Looks like there are outliers

```python
# Survival rate by embarkation port
sns.barplot(data=df, x="embarked", y="survived", palette="pastel")
plt.title("Survival Rate by Embarkation Port")
plt.show()
```

```python
# Survival by 'who' category
sns.countplot(data=df, x="who", hue="survived", palette="viridis")
plt.title("Survival by Passenger Type (Men, Women, Children)")
plt.xticks(ticks=["man","woman","child"],labels=[1,2,3])
plt.show()
```

```python
# Boxplot for age and fare
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=df, y="age", ax=axes[0], palette="Set2")
axes[0].set_title("Outliers in Age")
sns.boxplot(data=df, y="fare", ax=axes[1], palette="Set2")
axes[1].set_title("Outliers in Fare")
plt.tight_layout()
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=["object", "category"]).columns
data_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# 2. Calculate correlation matrix
correlation_matrix = data_encoded.corr()

# 3. Identify redundant columns (correlation > threshold)
threshold = 0.8
redundant_columns = set()

for col in correlation_matrix.columns:
    for row in correlation_matrix.index:
        if col != row and abs(correlation_matrix.loc[row, col]) > threshold:
            redundant_columns.add(col)

# 4. Drop redundant columns
data_cleaned = data_encoded.drop(columns=redundant_columns)

# Print results
print("Original Columns:", list(df.columns))
print("Redundant Columns:", redundant_columns)
print("Cleaned Dataset:\n", data_cleaned)
```
